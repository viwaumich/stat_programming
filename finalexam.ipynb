```r
i = 0
upper = 0
lower = 0
my_median <- function(v){
  sorted = sort(v)
  if(length(sorted) %% 2 == 0){
    lower = length(v)/2
    upper = length(v)/2+1
    i = mean(lower, upper)
    return (v[i])
  }
  if(length(sorted) %% 2 == 1){
    i = length(v)/2 + 0.5
    return (v[i])
  }
}
```
